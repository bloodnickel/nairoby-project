# Timeline/Evolution of Diaspora communities -- Demographic Trends 2000s to Today

Germany: From "Gastarbeiter" to Humanitarian Hub
    2000s: The landscape was dominated by the "Legacy" migration mix of the 60s/70s - primarily Turkey, Italy, and Greece.
    2004-2011: EU Expansion brought a massive influx of Polish and Romanian workers.
    2015-Present: The "Refugee Crisis" and the 2022 war in Ukraine fundamentally reshuffled Germany’s demographics. Syria and Ukraine skyrocketed into top 3, displacing older Mediterranean and Maghreb communities in terms of sheer growth.

United Kingdom: The Brexit Pivot
    2000s: High migration from the Commonwealth (India, Pakistan, Nigeria).
    2004: The "Polish Surge." Following EU expansion, the Polish community in the UK went from roughly 75,000 to over 1 million at its peak.
    Post-Brexit (2020-Today): We see a "re-sorting." The Polish population is actually declining as many return home or move to DE. Meanwhile, non-EU migration (India, Nigeria, Philippines) has surged to fill labor gaps in health and care sectors.

Poland: The Donbas Transformation
    2000s: Almost zero net immigration; Poland was a country people left.
    2014-Today: The "Great Transformation." Since the Donbas conflict began, and especially since 2022, Poland has transitioned from one of the most ethnically homogeneous countries in Europe to a major host nation.
    New Faces: Beyond Ukrainians, you now see growing Belarusian (political) and Central Asian/Indian (work visa) communities that were virtually non-existent in the 2000 census.
