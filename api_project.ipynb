```python
#Poland part
import pandas as pd
from pathlib import Path

p = Path('.')

files = [p / f"{year}_Poland.json" for year in range(2020, 2026)]

dfs = []
for f in files:
    j = pd.read_json(f, orient='records')
    if isinstance(j, pd.DataFrame) and j.empty:
        import json
        with open(f, 'r', encoding='utf-8') as fh:
            data = json.load(fh)
        for key in ('dataset', 'results', 'data', 'items'):
            if key in data and isinstance(data[key], list):
                j = pd.DataFrame(data[key])
                break
        else:
            if isinstance(data, list):
                j = pd.DataFrame(data)
            else:
                j = pd.DataFrame([data])
    dfs.append(j)

df = pd.concat(dfs, ignore_index=True, sort=False)

if 'year' not in df.columns:
    years = []
    for f in files:
        try:
            years.append(int(f.stem.split('_')[0]))
        except:
            years.append(None)
    blocks = [len(d) for d in dfs]
    year_col = []
    for y, cnt in zip(years, blocks):
        year_col += [y] * cnt
    if len(year_col) == len(df):
        df['year'] = year_col

print(df.shape)
df.to_csv('Poland_2020_2025_combined.csv', index=False, encoding='utf-8')
```

```python
#Find general kg AND money value for Partner country

agg = hot_df.groupby('partnerDesc', dropna=True).agg(
    sum_fobvalue=pd.NamedAgg(column='fobvalue', aggfunc='sum'),
    sum_qty=pd.NamedAgg(column='qty', aggfunc='sum')
).reset_index()

agg['sum_fobvalue_fmt'] = agg['sum_fobvalue'].map(lambda x: f"{x:.2f}")
agg['sum_qty_fmt'] = agg['sum_qty'].map(lambda x: f"{x:.2f}")

result_df = agg[['partnerDesc', 'sum_fobvalue_fmt', 'sum_qty_fmt']].rename(
    columns={
        'partnerDesc': 'Import Country',
        'sum_fobvalue_fmt': 'Total_USD',
        'sum_qty_fmt': 'Total_Qty_kg'
    }
)

print(result_df)
```

```python
#Find general kg AND money value for Reporter country
agg = hot_df.groupby('reporterDesc', dropna=True).agg(
    sum_fobvalue=pd.NamedAgg(column='fobvalue', aggfunc='sum'),
    sum_qty=pd.NamedAgg(column='qty', aggfunc='sum')
).reset_index()

agg = agg[agg['sum_fobvalue'] != 0].reset_index()

agg['sum_fobvalue_fmt'] = agg['sum_fobvalue'].map(lambda x: f"{x:.2f}")
agg['sum_qty_fmt'] = agg['sum_qty'].map(lambda x: f"{x:.2f}")

result_df = agg[['reporterDesc', 'sum_fobvalue_fmt', 'sum_qty_fmt']].rename(
    columns={
        'reporterDesc': 'Import Country',
        'sum_fobvalue_fmt': 'Total_USD',
        'sum_qty_fmt': 'Total_Qty_kg'
    }
)

print(result_df)
```
