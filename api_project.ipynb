```python
#Poland part
import pandas as pd
from pathlib import Path
pd.options.display.float_format = '{:,.2f}'.format

p = Path(r"../nairoby-project/UN_tradeDB")

files = [p / f"{year}_Poland.json" for year in range(2020, 2026)]

dfs = []
for f in files:
    j = pd.read_json(f, orient='records')
    if isinstance(j, pd.DataFrame) and j.empty:
        import json
        with open(f, 'r', encoding='utf-8') as fh:
            data = json.load(fh)
        for key in ('dataset', 'results', 'data', 'items'):
            if key in data and isinstance(data[key], list):
                j = pd.DataFrame(data[key])
                break
        else:
            if isinstance(data, list):
                j = pd.DataFrame(data)
            else:
                j = pd.DataFrame([data])
    dfs.append(j)

merge_hot_df1_pepper = pd.concat(dfs, ignore_index=True, sort=False)

if 'year' not in merge_hot_df1_pepper.columns:
    years = []
    for f in files:
        try:
            years.append(int(f.stem.split('_')[0]))
        except:
            years.append(None)
    blocks = [len(d) for d in dfs]
    year_col = []
    for y, cnt in zip(years, blocks):
        year_col += [y] * cnt
    if len(year_col) == len(merge_hot_df1_pepper):
        merge_hot_df1_pepper['year'] = year_col

print(merge_hot_df1_pepper.shape)
merge_hot_df1_pepper.to_csv('Poland_2020_2025_combined.csv', index=False, encoding='utf-8')
```

```python
# General Funtion with print of Import and Top 10
for year in range (2020,2026):
    y = merge_hot_df1_pepper[merge_hot_df1_pepper["year"] == year].copy()
    countries = merge_hot_df1_pepper.partnerDesc.dropna().unique().tolist()
    for country in countries:
        yy = y[y["partnerDesc"] == country]
        if yy.empty:
            continue
        g = (y.groupby(["Country", "Typical pepper_normalized"], as_index=False).agg({"Percent": "sum","alloc_fobvalue": "sum","alloc_qty": "sum"}).sort_values(["Country", "alloc_fobvalue"], ascending=[True, False]))
        g = g.rename(columns={"Typical pepper_normalized": "Most common pepper","Percent":"Percent","alloc_fobvalue":"Import in $","alloc_qty":"Import in kg"})
        filename = f"Import in {country} in {year} (with pepper names).xlsx"
        g.to_excel(filename, index=False, engine="openpyxl")
        money_by_pepper = g.groupby("Most common pepper", as_index=False)[["Import in kg", "Import in $"]].sum().sort_values("Import in $", ascending=False)    
        top10 = money_by_pepper.reset_index().head(10)
        filename1 = f"Top 10 Peppers Import in {country} in {year} (with pepper names).xlsx"
        top10.to_excel(filename1, index=False, engine="openpyxl")
        print(f"Top 10 in {country} in {year}:")
        print (top10)
```

```python
#preparing for visualisation top 3 by money and kg

start_year, end_year = 2020, 2025 
countries = merge_hot_df1_pepper.partnerDesc.unique().tolist()

for country in countries:
    rows = []
    for year in range(start_year, end_year + 1):
        yy = merge_hot_df1_pepper[(merge_hot_df1_pepper["year"] == year) & (merge_hot_df1_pepper["partnerDesc"] == country)].copy()
        if yy.empty:
            continue
        g = (yy.groupby("Typical pepper_normalized", as_index=False)
               .agg({"alloc_fobvalue": "sum", "alloc_qty": "sum"})
               .rename(columns={
                   "Typical pepper_normalized": "Most common pepper",
                   "alloc_fobvalue": "Import in $",
                   "alloc_qty": "Import in kg"
               })
               .sort_values("Import in $", ascending=False))
        top3 = g.head(3).copy()
        if top3.empty:
            continue
        top3["Year"] = year
        top3["Country"] = country
        top3 = top3[["Country", "Year", "Most common pepper", "Import in $", "Import in kg"]]
        rows.append(top3)
    if not rows:
        continue
    top3 = pd.concat(rows, ignore_index=True)
    filename = f"Top3_peppers_by_money_{country}_2020-2025.xlsx"
    top3.to_excel(filename, index=False, engine="openpyxl")
    print(f"Saved {filename} ({len(top3)} rows)")
```

```python
#preparing plots 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

merge_hot_df1_pepper = pd.read_excel('Top3_peppers_by_money_Germany_2020-2025.xlsx', engine='openpyxl')

def parse_number(s):
    if pd.isna(s): return None
    s = str(s).strip().replace(' ', '')
    if ',' in s and '.' in s:
        s = s.replace('.', '').replace(',', '.')
    elif ',' in s:
        s = s.replace(',', '.')
    s = ''.join(ch for ch in s if (ch.isdigit() or ch == '.' or ch == '-'))
    try: return float(s) if s not in ('', '.', '-') else None
    except: return None

for col in ["Import in $", "Import in kg"]:
    if col in merge_hot_df1_pepper.columns:
        merge_hot_df1_pepper[col] = merge_hot_df1_pepper[col].apply(parse_number)

merge_hot_df1_pepper["Year"] = pd.to_numeric(merge_hot_df1_pepper["Year"], errors='coerce')

records = []
for year, grp in merge_hot_df1_pepper.groupby("Year"):
    top_peppers = grp.groupby("Most common pepper")["Import in $"].sum().nlargest(3).index
    sel = grp[grp["Most common pepper"].isin(top_peppers)].groupby(["Year","Most common pepper"]).agg({
        "Import in $":"sum","Import in kg":"sum"
    }).reset_index()
    records.append(sel)
df_top_per_year = pd.concat(records, ignore_index=True)

df_melt_money = df_top_per_year[['Year','Most common pepper','Import in $']].rename(columns={'Import in $':'value'})
df_melt_money['value'] = df_melt_money['value'] / 1e6
df_melt_money['metric'] = 'money'
df_melt_kg = df_top_per_year[['Year','Most common pepper','Import in kg']].rename(columns={'Import in kg':'value'})
df_melt_kg['value'] = df_melt_kg['value'] / 1e6
df_melt_kg['metric'] = 'kg'
df_plot = pd.concat([df_melt_money, df_melt_kg], ignore_index=True)

sns.set(style="whitegrid")
fig, axes = plt.subplots(1,2, figsize=(16,6))


sns.barplot(data=df_plot[df_plot['metric']=='money'], x='Year', y='value', hue='Most common pepper', ax=axes[0], ci=None)
axes[0].set_title('Import ($) — top 3 pepper types in years 2020-2025')
axes[0].set_ylabel('Import in $ (millions)')
axes[0].yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, pos: f'{x:,.1f}M'))

# Кг
sns.barplot(data=df_plot[df_plot['metric']=='kg'], x='Year', y='value', hue='Most common pepper', ax=axes[1], ci=None)
axes[1].set_title('Import (kg) — top 3 pepper types in years 2020-2025')
axes[1].set_ylabel('Import in thousands tonn')
axes[1].yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, pos: f'{x:,.1f}M'))

for ax in axes:
    ax.legend(title='Pepper', bbox_to_anchor=(1.02,1), loc='upper left')

plt.tight_layout()

fig.savefig('Germany peppers_imports_2020-2025.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Checking change in export tendency by country by year dropping main reexporters and producers
import os
from matplotlib.ticker import FuncFormatter

os.makedirs('plots', exist_ok=True)

drop_countries = ['Netherlands', 'Spain', 'Türkiye', 'Morocco', 'France']
main_filtered = main[~main['Country'].isin(drop_countries)].reset_index(drop=True)

for col in ['alloc_fobvalue','alloc_qty','fobvalue','qty']:
    if col in main_filtered.columns:
        main_filtered[col] = (main_filtered[col].astype(str)
                                           .str.replace(r'[,\s]', '', regex=True)
                                           .replace('', '0')
                                           .astype(float))
main_filtered['alloc_fobvalue_m'] = main_filtered['alloc_fobvalue'] / 1_000_000.0
if 'year' in main_filtered.columns:
    main_filtered['year'] = main_filtered['year'].astype(int)

def millions(x, pos):
    return f'{x:.1f}M'
formatter = FuncFormatter(millions)


top_k = 10          # change for countries number to show
figsize_single = (10, 6) 

partners = main_filtered['partnerDesc'].unique()

for p in partners:
    sub = main_filtered[main_filtered['partnerDesc'] == p]
    if sub.empty:
        continue
    top_countries = (sub.groupby('Country', as_index=False)['alloc_fobvalue']
                        .sum()
                        .nlargest(top_k, 'alloc_fobvalue')['Country'].tolist())
    sub_top = sub[sub['Country'].isin(top_countries)].sort_values('year')

    fig, ax = plt.subplots(figsize=figsize_single)
    sns.lineplot(data=sub_top, x='year', y='alloc_fobvalue_m', hue='Country',
                 marker='o', ax=ax, palette='tab10')

    ax.set_title(p)
    ax.set_ylabel('Import in mil $')
    ax.set_xlabel('Year')
    ax.yaxis.set_major_formatter(formatter)
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    # move legend out of main picture
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.75, box.height])  # reshape to get place for leg
    ax.legend(title='Country', bbox_to_anchor=(1.02, 0.5), loc='center left', fontsize='small')
  
    filename = f'plots/{p.replace("/", "_")}_imports.png'
    os.makedirs('plots', exist_ok=True)
    fig.savefig(filename, bbox_inches='tight', dpi=300)
   
    plt.show()
```
